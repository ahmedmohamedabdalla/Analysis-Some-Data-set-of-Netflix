```python
#Data Visualization
v1=['TLS','No TV Shows','No Movies','Cost Basic $    ', 'Cost Standard $']
d=list([df['Total Library Size'].sum(),df['No. of TV Shows'].sum(),df['No. of Movies'].sum(),df['Cost Per Month - Basic ($)'].sum(),
        df['Cost Per Month - Standard ($)'].sum()])
plt.bar(v1,d,color='#32AAAA',width=.5)

plt.show()
```

```python
a=df['Cost Per Month - Basic ($)']
b=df['Cost Per Month - Standard ($)']
c=df['Cost Per Month - Premium ($)']
d=df['No. of Movies']
plt.scatter(d,a,color='#253F58')
plt.scatter(d,b,color='#D87151')
plt.scatter(d,c,color='#DEDE0E')
plt.show()
```

```python
d=df['Cost Per Month - Basic ($)'].sum()
d1=df['Cost Per Month - Standard ($)'].sum()
d2=df['Cost Per Month - Premium ($)'].sum()
size=list([d,d1,d2])
ex=[0.1,0.1,0.1]
c=['#F7C45F','#727481','lightskyblue']
la=['Basic','Standard','Premium']
fig,ax=plt.subplots()
ax.pie(size,labels=la,colors=c,shadow=True,startangle=90,explode=ex)
ax.set_title('Sum Of Cost')
plt.show()
```

```python
a1=df['No. of TV Shows']
a2=df['No. of Movies']
plt.subplot(1,2,1)
plt.plot(a1,color='#D65352',linestyle='-',linewidth='2')
plt.xlabel('TV Shows')
plt.ylabel('Number')
plt.subplot(1,2,2)
plt.plot(a2,color='#28364A',linestyle='-',linewidth='2')
plt.xlabel('Movies')
plt.ylabel('Number')
plt.tight_layout(2)
plt.show()
```
